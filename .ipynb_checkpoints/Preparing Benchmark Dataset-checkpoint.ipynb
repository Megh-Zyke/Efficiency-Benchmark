```python
import pandas as pd 
import numpy as np 
```

```python
db = pd.read_excel("temp_benchmark_data.xlsx")
db = db.iloc[:22,:]

db.head()
```

```python
columns = ["meta_data","topic" , "difficulty", "question", "problem_description" , "entry_point"]
dataset = pd.DataFrame(columns = columns)
dataset.head()
```

```python
def topic():
    
    topics = [ "Array" , "Greedy" , "Dynamic Programming" , "Backtracking", "Divide and Conquer" , "DFS" , "BFS" , "Linked Lists" , "Two Pointers" , "Sliding Window", "Binary Trees" , 
              "Bit manipulation" , "Sorting" , "Mathematical Reasoning" , "Graph"]
    for i in range(len(topics)):
        print(f"{i}: {topics[i]}")
    print("Enter the topic index:")
    n = input()
    n = n.split(" ")
    return [topics[int(i)] for i in n] 

```

```python
def metaDataGen(index : int):
    meta = {}
    entities = db.iloc[index, :]
    meta["index"] = entities[0]
    meta["question name"] = entities[1]
    meta["difficulty"] = entities[3]
    meta["entry point"] = entities[5]
    meta["topic"] = topic()
    return meta
    
```

```python
##### for i in range(17,22):
    entities = db.iloc[i, :]
    print(entities[1])
    data = metaDataGen(i)   
    print(data)
    row_data = pd.Series({
        columns[0]: str(data),  # Store full dict as a string
        columns[1]: str(data.get("topic", None)),  # Assign topic safely
        columns[2]: data.get("difficulty", None),  # Assign difficulty safely
        columns[3]: data.get("question name", None),  # Assign question name safely
        columns[4]: entities[2],  # Ensure correct column name
        columns[5]: data.get("entry point", None)  # Assign entry point safely
    })
    print(row_data)
    dataset.loc[i] = row_data
```
